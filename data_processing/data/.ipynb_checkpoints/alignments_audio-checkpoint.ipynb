```python
for subdirs, dirs, files in os.walk(alignment_path):
    f_count = 0
    
    for f in files:
        split_f = f.split('_')
        ppno = split_f[1]
        img_id = split_f[2].split('.')[0]
        
        f_path = os.path.join(alignment_path, f)
        print(f_path)
        print()
        with open(f_path, 'r') as map_file:
            mapped = json.load(map_file)
            
            for fr in mapped['fragments']:
                print(fr['lines'][0], fr['begin'], fr['end']) 
         
        f_count += 1
        print()
        if f_count == 3:
            break
```
